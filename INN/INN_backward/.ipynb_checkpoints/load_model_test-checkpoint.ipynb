```python
%matplotlib inline
import torch
import numpy as np

pretrained_net = torch.load('inn_model_state6.pkl',map_location=lambda storage, loc: storage)

for key, v in enumerate(pretrained_net):
    print(key, v)
    
import model
    
model2 = model.model
print(model2)

model2.load_state_dict(pretrained_net)
```

```python
## plot the results
from numpy import sin, cos
import matplotlib.pyplot as plt
device = 'cuda' if torch.cuda.is_available() else 'cpu'
import config as c

x_low = -1
x_high = 1
ns_dim = 100
x_dim = 2
f = 3

# plot the contour of sinewave function
delta = 0.02
x = np.arange(-1.0, 1.0, delta)
y = np.arange(-1.0, 1.0, delta)
X, Y = np.meshgrid(x, y)
Z = sin(f * np.pi * X) + cos(f * np.pi * Y)

xx = []
for i in range(x_dim):
    xx.append(np.linspace(x_low, x_high, num=ns_dim))

x = np.array(np.meshgrid(*xx))
data_x = np.concatenate([np.reshape(np.ravel(x[i, :]), [-1, 1]) for i in range(2)], axis=1)
x_test = torch.tensor(data_x, dtype=torch.float)

y_pred = model2(torch.cat((x_test, torch.zeros(10000, 2)), dim=1).to(device)).data
y_test = np.reshape(y_pred[:,  -c.ndim_y].cpu().numpy(), (100, 100))

fig, ax = plt.subplots(figsize=(10,10))
CS1 = ax.contour(X, Y, y_test)
ax.clabel(CS1, inline=True, fontsize=10)
ax.set_title('Simplest default with labels')
name = './figure/predict' + 'epoch' + '.png'
fig.savefig(name, dpi=None)

# samples
print('generating new samples')
fig, ax = plt.subplots(figsize=(10,10))
# CS = ax.contour(X, Y, Z)
CS1 = ax.contour(X, Y, y_test)
ax.clabel(CS1, inline=True, fontsize=10)
ax.set_title('Simplest default with labels')

n_samps = 50

y_fix = np.zeros((n_samps, 1)) - 1.2
y_fix = torch.tensor(y_fix, dtype=torch.float)

# y_fix += c.add_y_noise * torch.randn(n_samps, c.ndim_y)
y_fix = torch.cat([torch.randn(n_samps, c.ndim_z),
                   c.add_z_noise * torch.zeros(n_samps, 0), y_fix], dim=1)
y_fix = y_fix.to(device)

rev_x0 = model.model(y_fix, rev=True)
rev_x = rev_x0.cpu().data.numpy()
np.shape(rev_x)
ax.plot(rev_x[:, 0], rev_x[:, 1], 'xr')
plt.xlim([-1, 1])
plt.ylim([-1, 1])
# name1 = './figure/samples' + str(epoch) + '.png'
# fig.savefig(name1, dpi=None)
```
